# Day 1 - Data Inspection & Validation

Analyse datasets, explore fund master and validate AMFI codes.

```python
import pandas as pd
from pathlib import Path

RAW_DIR = Path("../data/raw")

CSV_FILES = [
    "01_fund_master.csv",
    "02_nav_history.csv",
    "03_aum_by_fund_house.csv",
    "04_monthly_sip_inflows.csv",
    "05_category_inflows.csv",
    "06_industry_folio_count.csv",
    "07_scheme_performance.csv",
    "08_investor_transactions.csv",
    "09_portfolio_holdings.csv",
    "10_benchmark_indices.csv"
]

dataframes = {}

for file in CSV_FILES:
    df = pd.read_csv(RAW_DIR / file)
    dataframes[file] = df

    print(f"\n{file}")
    print("Shape:", df.shape)
    print(df.dtypes)
    display(df.head())
```

## Explore Fund Master

Review fund houses, categories, sub-categories and risk grades.

```python
fund_master = dataframes["01_fund_master.csv"]

print("Fund Houses")
print(fund_master["fund_house"].unique())

print("\nCategories")
print(fund_master["category"].unique())

print("\nSub Categories")
print(fund_master["sub_category"].unique())

print("\nRisk Grades")
print(fund_master["risk_category"].unique())
```

## Validate AMFI Codes

```python
nav_history = dataframes["02_nav_history.csv"]

master_codes = set(fund_master["amfi_code"])
nav_codes = set(nav_history["amfi_code"])

missing_codes = master_codes - nav_codes

print("Fund Master Codes:", len(master_codes))
print("NAV History Codes:", len(nav_codes))
print("Missing Codes:", len(missing_codes))
```

## Data Quality Summary

```python
fund_master = dataframes["01_fund_master.csv"]
nav_history = dataframes["02_nav_history.csv"]

master_codes = set(fund_master["amfi_code"])
nav_codes = set(nav_history["amfi_code"])

missing_codes = len(master_codes - nav_codes)

total_nulls = sum(df.isnull().sum().sum() for df in dataframes.values())
total_duplicates = sum(df.duplicated().sum() for df in dataframes.values())

summary = f"""
Day 1 Data Quality Summary
==========================

Datasets Loaded: {len(dataframes)}
Total Null Values: {total_nulls}
Total Duplicate Rows: {total_duplicates}

Fund Master Codes: {len(master_codes)}
NAV History Codes: {len(nav_codes)}
Missing AMFI Codes: {missing_codes}

Conclusion:
All datasets were loaded and inspected successfully.
AMFI code validation completed.
Further cleaning will be performed during data transformation.
"""

with open("../reports/data_quality_day1.txt", "w") as f:
    f.write(summary)

print(summary)
print("\nSaved: reports/data_quality_day1.txt")
```
